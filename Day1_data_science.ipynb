```
 
 
import random
str1="OBANWRI"
l=list(str1)
str2="RAINBOW"
s=list(str2)
while s!=l:
    random.shuffle(l)
    print("".join(l))
print("\n\nYou Guess The Correct Word!!\n")
print("".join(l))
```

```
 
 
cp=eval(input("Enter cost price: "))
sp=eval(input("Enter selling price: "))
if cp>sp:
    print("\nLOSS")
elif cp<sp:
    print("\nPROFIT")
else:
    print("\nNEITHER")
```
